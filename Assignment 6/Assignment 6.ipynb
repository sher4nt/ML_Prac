```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, RepeatVector, Reshape, Lambda
```

```python
model = tf.keras.models.Sequential()

model.add(Dense(units=20, input_dim=10, kernel_initializer=tf.keras.initializers.Identity()))
model.add(RepeatVector(2))
model.add(Flatten())
model.add(Dropout(rate=0.2))
model.add(Activation('relu'))
model.add(Reshape((8,5)))
model.add(Dense(units=10, kernel_initializer=tf.keras.initializers.Identity()))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Lambda(tf.keras.backend.exp, output_shape=80))
model.add(Dropout(rate=0.1))
model.add(Dense(units=3, kernel_initializer=tf.keras.initializers.Identity()))
model.add(Activation('softmax'))
model.add(Lambda(lambda x: tf.keras.backend.max(x, axis=-1), output_shape=1))
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
model.compile(loss='mean_squared_error', optimizer = 'adam')
```

```python
data = np.arange(10).reshape((1,-1))
prediction = model.predict(data)
print(prediction)
```
