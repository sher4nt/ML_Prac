```python
import numpy as np
import pandas as pd


from statsmodels.api import OLS
from statsmodels.tools.tools import add_constant
from sklearn.linear_model import Lasso, LassoCV
```

```python
data = pd.read_csv("test_sample.csv")
data
```

```python
X = data.drop(['Y'], axis=1)
Y = data['Y']
```

```python
regressorCV = LassoCV(cv=5, random_state=1)
regressorCV.fit(X, Y)
```

```python
alpha_min = regressorCV.alpha_
```

```python
regressor_lasso = Lasso(alpha=alpha_min, random_state=1)
regressor_lasso.fit(X, Y)
```

```python
coefs = regressor_lasso.coef_
lasso_elim = np.argwhere(coefs==0)
```

```python
regressor = OLS(Y, add_constant(X)).fit()
pvalues = regressor.pvalues.to_numpy()
ols_elim = np.argwhere(pvalues > 0.1) - 1
```

```python
lasso_zeros = ' '.join([str(idx[0]) for idx in lasso_elim])
lm_zeros = ' '.join([str(idx[0]) for idx in ols_elim])
pd.DataFrame([lasso_zeros,lm_zeros], index = ['eliminated_by_Lasso','eliminated_by_lm']).to_csv('answer.csv')
```
