```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.inspection import permutation_importance

from scikeras.wrappers import KerasRegressor
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
data = pd.read_csv("data.csv")
print(data.shape)
data.head()
```

```python
x = data.drop('y', axis=1)
y = data['y']
```

```python
linear_model = LinearRegression()
linear_model.fit(x, y)
print(mean_absolute_error(y, linear_model.predict(x)))
```

```python
r = permutation_importance(linear_model, x, y,
                           n_repeats=30,
                           random_state=0)
```

```python
importances = r.importances_mean
importance_dict = {}
for i in range(importances.shape[0]):
    importance_dict[i + 1] = importances[i]
sorted(importance_dict, key=importance_dict.get, reverse=True)[:5]
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(200, input_dim=10, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer = 'adam')
    return model
```

```python
neural = KerasRegressor(build_fn=create_model, epochs=30, batch_size=128,verbose=0)
neural.fit(x, y)
```

```python
r = permutation_importance(neural, x, y,
                           n_repeats=30,
                           random_state=0)
```

```python
importances = r.importances_mean
importance_dict = {}
for i in range(importances.shape[0]):
    importance_dict[i + 1] = importances[i]
sorted(importance_dict, key=importance_dict.get, reverse=True)[:5]
```
