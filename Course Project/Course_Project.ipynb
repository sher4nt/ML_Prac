```python
import numpy as np
import pandas as pd
import xgboost
```

```python
train_data = pd.read_csv("train_sample.csv")
test_data = pd.read_csv("test_sample.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
X_train, y_train = train_data.drop('target', axis=1).values, train_data['target'].values
X_test = test_data.drop('id', axis=1).values
```

```python
params = {'objective': 'multi:softprob', 'eval_metric': 'mlogloss', 'num_class': 9}
train_xgb = xgboost.DMatrix(X_train, y_train)
test_xgb  = xgboost.DMatrix(X_test)
trainRound = 200
```

```python
model = xgboost.train(params, train_xgb, trainRound)
```

```python
pred = model.predict(test_xgb)
pred
```

```python
sub_data = pd.DataFrame(pred, index=test_data.id.values)
sub_data.index.name = 'id'
sub_data
```

```python
sub_data.to_csv('submission.csv')
```
