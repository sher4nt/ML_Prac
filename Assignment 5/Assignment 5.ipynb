```python
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
train = pd.read_csv("train_sample.csv")
test = pd.read_csv("test_sample.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
X_train, y_train = train.drop(['ID', 'class'], axis=1), train.drop('ID', axis=1)['class']
X_test = test.drop('ID', axis=1)
```

```python
pipe = Pipeline([('scaler', MinMaxScaler()), ('svc', SVC())])
```

```python
cv = GridSearchCV(
    estimator=pipe,
    param_grid={
        'svc__gamma':.1**np.arange(-1,4),
        'svc__C': 5*np.arange(3,6)
    }
)

cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
model = pipe
model.C = 15
model.gamma = 10
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
test['class'] = y_pred
test[['ID','class']].to_csv('submission.csv', index=False)
```
